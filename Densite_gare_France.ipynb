```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'liste-des-gares.csv'
gares_df = pd.read_csv(file_path, delimiter=';')


latitude = gares_df['y_wgs84']
longitude = gares_df['x_wgs84']


plt.figure(figsize=(12, 10))  
plt.hexbin(longitude, latitude, gridsize=50, cmap='RdYlGn_r', mincnt=1, vmin=1, vmax=20)


cb = plt.colorbar(label='Number of points')

plt.title('Gares fran√ßaises - Carte de densit√©')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)

plt.savefig('Gare_France.pdf')
from IPython.display import FileLink

display(FileLink('Gare_France.pdf'))
plt.show()  

```
